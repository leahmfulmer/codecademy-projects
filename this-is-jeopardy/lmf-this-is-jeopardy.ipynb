2. Write a function that filters the dataset for questions that contains all of the words in a list of words. For example, when the list `["King", "England"]` was passed to our function, the function returned a DataFrame of 49 rows. Every row had the strings `"King"` and `"England"` somewhere in its `" Question"`.

   Test your function by printing out the column containing the question of each row of the dataset.

````python
# Define a function to filter rows by items in a wordlist
# This function has already been changed for case fluency

def filter_wordlist(wordlist):
    questions_wordlist = []
    answers_wordlist = []
    
    for (i, question) in enumerate(jeopardy["Question"]):
        if all(word.lower() in question.lower() for word in wordlist):
            questions_wordlist.append(question)
            answers_wordlist.append(jeopardy["Answer"][i])

    df_wordlist = pd.DataFrame({'Questions with "'+', '.join(wordlist)+'"': \
                                questions_wordlist, 'Answers with "'+', '.join(wordlist)+'"': answers_wordlist})
    return df_wordlist

example = filter_wordlist(["King", "England"])
print("The output has {} rows.".format(example.shape[0]))
example.head()
````

````python
# I notice that my filter function is much less effecient than the Solution.
# I'd like to learn how to make mine more efficint, too, especially with lambda functions.

def filter_data_wordlist(data, wordlist):
    
    filter = lambda x: all(word.lower() in x.lower() for word in wordlist)
    
    return data.loc[data["Question"].apply(filter)]

example = filter_data_wordlist(jeopardy, ["King", "England"])
print("The output has {} rows.".format(example.shape[0]))
example.head()
````

````python
# Apply filter to my interests

my_interests = filter_data_wordlist(jeopardy, ["Hindu"])
print("The output has {} rows.".format(my_interests.shape[0]))
my_interests.head()
````

````python
# Funily enough, every question about Hindu rivers refers to the Ganges.
# This makes some sense, because it is considered the most sacred of rivers.

hindu_river = filter_data_wordlist(jeopardy, ["Hindu", "river"])
print("The output has {} rows.".format(hindu_river.shape[0]))
hindu_river

# If you're ever asked about a Hindu river on Jeopardy, you know what to do...
````

4. We may want to eventually compute aggregate statistics, like `.mean()` on the `" Value"` column. But right now, the values in that column are strings. Convert the`" Value"` column to floats. If you'd like to, you can create a new column with float values.

   While most of the values in the `" Value"` column represent a dollar amount as a string, note that some do not &mdash; these values will need to be handled differently!

   Now that you can filter the dataset of question, use your new column that contains the float values of each question to find the "difficulty" of certain topics. For example, what is the average value of questions that contain the word `"King"`?
   
   Make sure to use the dataset that contains the float values as the dataset you use in your filtering function.

````python
# Create a column for Float Value

jeopardy["Float Value"] = jeopardy.apply \
(lambda row: np.nan if row["Value"] == 'no value' \
 else float(row["Value"].strip('$').replace(',','')), axis=1)
jeopardy[-5:]
````

````python
# Note: Wherever jeopardy["Value"] == 'no value' it indicates a Final Jeopary! question
final = jeopardy[jeopardy["Value"] == 'no value']
final.head()
````

````python
# Compute the mean Value for all questions, excluding Final Jeopardy! questions
print("The mean value for a Jeopardy! question is {} points.".format(round(jeopardy["Float Value"].mean())))
````

````python
# Create a function that returns the average value of questions that contain a given wordlist

def avg_value_filter(wordlist):
    value_wordlist = []
    
    for (i, question) in enumerate(jeopardy["Question"]):
        if all(i.lower() in question.lower() for i in wordlist):
            value_wordlist.append(jeopardy["Float Value"][i])

    avg_value = round(np.nanmean(value_wordlist))
    print('The mean value of questions that contain "'+', '\
          .join(wordlist)+'"'+' is {} points.'.format(avg_value))
    
    return avg_value

avg_value_filter(["King"])
avg_value_filter(["Hindu"])
````

````python
# Instead of writing an entirely new function, 
# try it again using the function that already exists

king_data = filter_data_wordlist(jeopardy, ["King"])
print("The mean value for questions with the word 'King' is {} points."\
      .format(round(king_data["Float Value"].mean())))

hindu_data = filter_data_wordlist(jeopardy, ["Hindu"])
print("The mean value for questions with the word 'Hindu' is {} points."\
      .format(round(hindu_data["Float Value"].mean())))                       
````

5. Write a function that returns the count of unique answers to all of the questions in a dataset. For example, after filtering the entire dataset to only questions containing the word `"King"`, we could then find all of the unique answers to those questions. The answer "Henry VIII" appeared 55 times and was the most common answer.

````python
# Create a function to return the count of unique answers after a filter
from collections import Counter

def unique_answers_filter(wordlist):
    answers_wordlist = []
    
    for (i, question) in enumerate(jeopardy["Question"]):
        if all(i.lower() in question.lower() for i in wordlist):
            answers_wordlist.append(jeopardy["Answer"][i])
            
    return Counter(answers_wordlist)

print(unique_answers_filter(["King"]).most_common(3))
print(unique_answers_filter(["Hindu"]).most_common(3))
````
